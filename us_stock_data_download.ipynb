```python
import pandas as pd
sp500 = pd.read_html("https://en.wikipedia.org/wiki/List_of_S%26P_500_companies")[0]
tickers = sp500.sort_values('GICS Sector', ascending=True)['Symbol'].head(300).tolist()
def yahoo_ticker_fix(ticker):
    # For .B, .A, .C, etc, convert to Yahoo format (e.g., BRK.B -> BRK-B)
    return ticker.replace('.', '-')

tickers = [yahoo_ticker_fix(ticker) for ticker in tickers]
```

```python
tickers
```

```python
import yfinance as yf
from datetime import date, timedelta

end   = date.today()
start = end - timedelta(days=365)      # last 1 year

raw = yf.download(
        tickers       = tickers,
        start         = start,
        end           = end,
        # group_by      = 'ticker',      # data comes back in a {ticker: DataFrame} layout
        threads       = True,
        auto_adjust   = False,         # keep raw (split-unadjusted) prices
        progress      = False)         # quiet mode
```

```python
import yfinance as yf
import pandas as pd

results = []
for tkr in tickers:
    t = yf.Ticker(tkr)
    info = t.info
    # Get latest daily price data (adjust period as needed)
    try:
        hist = t.history(period='30d')
        if hist.empty:
            print(f"Warning: no history for {tkr}")
            continue
        row = hist.iloc[-1]
        Date = row.name.date() if hasattr(row.name, 'date') else row.name
        price = row['Close']
        volume = row['Volume']
    except Exception as e:
        print(f"Error fetching price for {tkr}: {e}")
        continue

    # Fetch fields, with fallback to None if not available
    float_shares = info.get('floatShares')
    shares_out = info.get('sharesOutstanding')
    avg_volume = info.get('averageVolume')
    trailingPE = info.get('trailingPE')
    priceToBook = info.get('priceToBook')
    priceToSales = info.get('priceToSalesTrailing12Months')
    dividendYield = info.get('dividendYield')
    marketCap = info.get('marketCap')

    # Compose all variables
    results.append({
        'ticker': tkr,
        'trade_date': Date,
        'turnover_rate': volume / float_shares if float_shares else None,
        'turnover_rate_f': None,  # Not available for US stocks
        'volume_ratio': volume / avg_volume if avg_volume else None,
        'pe': trailingPE,
        'pe_ttm': trailingPE,  # trailing PE is the best yfinance has
        'pb': priceToBook,
        'ps': priceToSales,
        'ps_ttm': priceToSales,
        'dv_ratio': dividendYield,
        'dv_ttm': dividendYield,
        'total_share': shares_out,
        'float_share': float_shares,
        'free_share': None,  # Not available for US stocks
        'total_mv': marketCap,
        'circ_mv': price * float_shares if float_shares else None
    })

# Make DataFrame
df = pd.DataFrame(results)

# Print all variables, all columns
with pd.option_context('display.max_columns', None):
    print(df)
```
